```python
# Ejemplo Practico:
# En un lanzamiento de cuatro monedas, 
# cual es la probabilidad de obtener dos caras P(X=2), y 
# cual es la probabilidad de obtener 3 o menos caras P(X<=3) ..hint esto es = a sumar P(X=0)+P(X=1)+P(X=2)+P(X=3) 
# P(X=0) =  no sale ninguna cara
# P(X=1) = sale 1 cara
# P(X=2) = salen 2 caras
# P(X=3) = salen 3 caras

# espacio muestra (sample space) = 0,1,2,3,4
# Todas las muestras son independientes, pudiendolo modelar con una distribucion binomial.
# Y cada evento tiene la posibilidad de sacar cara o sello, 
# Debemos entonces, establecer la cantidad de eventos idependientes.  
n = 4
p = 0.5
results = [0,1,2,3,4] # No es buena practica escribir la lista a mano. Mejor generarla con: list(range(n+1))
freqs = [binom.pmf(i,n,p) for i in result] # Lista de Frecuencias, generada con comprenhension list.
# los argumentos de binom.pmf son nro de exitos, eventos (trials), y probabilidad
# Veremos con los siguiente, la probabilidad de que salga cada cara.
plt.bar(results, freqs)
for i, val in enumerate(freqs): # enumerate extrae el indice y el valor (en ese orden), de la coleccion, en este caso la lista. 
    plt.text(i,val,round(val,2),ha="center") # el primero y el segundo argumento son las coordenadas y el ultimo es el valor. Este ultimo lo redondeamos con la funcion round.El cuarto argumento es para centrar los valores en el grafico.
# Los argumentos de text son las coordenadas x,y, y el valor a escribir.
plt.grid(True)
plt.xlabel("Conteo de Caras")
plt.ylabel("Probabilidad")
plt.show()
```

```python
# Aca vemos que la probabilidad de obtener 2 caras es de 38%
```

```python
# Agregamos con list comprehension el acumulado
accum_freqs = [sum(freqs[:i+1]) for i in range(len(freqs))]
plt.bar(results, accum_freqs)
for i, val in enumerate(accum_freqs): # enumerate extrae el indice y el valor (en ese orden), de la coleccion, en este caso la lista. 
    plt.text(i,val,round(val,2),ha="center") 
plt.grid(True)
plt.xlabel("Conteo de Caras")
plt.ylabel("Probabilidad Acumulada")
plt.show()
```

```python
start = 0
end = 10
num = 100
numbers = [start+i*(end-start)/num for i in range(num)]
# La distribucion aca es continua ya que estoy generando un rango del 0 al 10 incluyendo todos los decimales.
# en el caso anterior son eventos discretos.

# los argumentos de norm.pdf son numero de sucesos, media, y desviacion estandar,(d por densidad porque es continua):
freqs = [norm.pdf(value,5,1) for value in numbers] 
# como ya no estoy trabajando en el espacio discreto, sino continuo, el for no nos sirve para graficar:
plt.plot(numbers, freqs) # Los argumentos de text son las coordenadas x,y, y el valor a escribir.
plt.grid(True)
plt.axvline(5, color="red") # graficamos la media y le cambiamos el color a la linea
plt.axvline(4, color="green") # Grafica una de desviacion estandar menos de la media: 5-1 = 4
plt.axvline(6, color="green") # Grafica una de desviacion estandar mas de la media 5+1 = 6
plt.axvline(3, color="purple") # Grafica 2 desviaciones estandar menos de la media: 5-2 = 3
plt.axvline(7, color="purple") # Grafica 2 desviaciones estandar mas de la media 5+2 = 7
plt.xlabel("Valores")
plt.ylabel("Probabilidad")
plt.show()

# La teoria dice que el 66% de las probabilidades estarian dentro de las lineas verdes.
# La teoria dice que la probabilidad de que este entre 3 y 7 es del 
# Como la Distribucion es Normal, la probabilidad es continua, entonces la tenemos que calcular por intervalo,
# y ya no por punto como en la distribucion binomial que es discreta.
```
